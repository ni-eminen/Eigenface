### Using timeit to measure runtime and practical time complexity

```python
import timeit
from eigenfaces_class import Eigenfaces

setsize_runtime = []
for dataset_size in range(100, 2500, 100):
    time = timeit.timeit(
    lambda: Eigenfaces(dataset_size).train_on_olivetti_dataset(),
    number=10
    )
    setsize_runtime.append((dataset_size, time * 1000 / 10))
```

```python
import matplotlib.pyplot as plt
obj = {}

obj['setsize'] = [i[0] for i in setsize_runtime]
obj['runtime'] = [i[1] for i in setsize_runtime]
```

### Plotting the runtime. x-axis depicts the size of the dataset (x images) and y is the runtime in milliseconds
Runtime increases exponentially with dataset size. Theoretically, the runtime of training the algorithm is O(nÂ³) where n is the amount of images in the dataset.

```python
plt.xlabel('Dataset size')
plt.ylabel('Runtime')
plt.plot('setsize', 'runtime', data=obj)
```
